```python
import matplotlib.pyplot as plt
import pandas as pd
# dendogram 
import scipy .cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
data = pd.read_csv("Mall_Customers.csv")
data.head()
```

```python
x = data.iloc[:,[3,4]].values
```

```python
# for dendogram
dendogram = sch.dendrogram(sch.linkage(x,method='ward'))
plt.title("Dendogram")
plt.xlabel("Customber")
plt.ylabel("Eucliden Distance")
plt.show()
```

```python
hc = AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_hc = hc.fit_predict(x)
```

```python
plt.scatter(x[y_hc == 0,0],x[y_hc == 0,1], s=50 , c ='red',label = 'Cluster-1')
plt.scatter(x[y_hc == 1,0],x[y_hc == 1,1],s = 50, c = 'green',label = 'Cluster-2')
plt.scatter(x[y_hc == 2,0],x[y_hc == 2,1], s=50 , c ='blue',label = 'Cluster-3')
plt.scatter(x[y_hc == 3,0],x[y_hc == 3,1],s = 50, c = 'black',label = 'Cluster-4')
plt.scatter(x[y_hc == 4,0],x[y_hc == 4,1], s=50 , c ='yellow',label = 'Cluster-5')
plt.title("Cluster of customers")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```
