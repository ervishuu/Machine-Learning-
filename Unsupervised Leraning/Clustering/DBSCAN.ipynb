```python
import pandas as pd
```

```python
data = pd.read_csv("Wholesale customers data.csv")
data.head(10)
```

```python
real_x = data.iloc[:,[1,2]].values
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
real_x = sc.fit_transform(real_x)
```

```python
dbscan = DBSCAN(eps = 0.5 , min_samples=15)
pred_dbscan = dbscan.fit(real_x)
print(pred_dbscan.labels_)
```

```python
label = pred_dbscan.labels_
size =30
plt.scatter(real_x[label ==-1,0],real_x[label ==-1,1],s = size , c = "red")
plt.scatter(real_x[label ==0,0],real_x[label ==0,1],s = size , c = "blue")
plt.xlabel("Cluster of Customers")
plt.ylabel("Milk")
plt.show()
```
