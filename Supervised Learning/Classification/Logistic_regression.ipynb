```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv("Ads.csv")
data.head(10)
```

```python
# split dataset
real_x = data.iloc[:,[2,3]].values
real_y  = data.iloc[:,4].values
```

```python
# divide testining and testing dataset

training_x ,test_x , training_y , test_y = train_test_split(real_x,real_y ,test_size = 0.25,random_state = 0) 
```

```python
# Feture Scalling btwn -1 to -2

scaler = StandardScaler()
training_x = scaler.fit_transform(training_x)
test_x =  scaler.fit_transform(test_x)
```

```python
classifer_LR = LogisticRegression(random_state=0)
classifer_LR.fit(training_x,training_y)
```

```python
# prediction

y_pred=classifer_LR.predict(test_x)
y_pred
```

```python
# checking 
test_y
```

```python
#confusion Matrics
c_m = confusion_matrix(test_y,y_pred)
c_m
# 63 and 25 correct prediction and total acciracy is 88 
```

```python
# traing dataset

from matplotlib.colors import ListedColormap
X_set , y_set = training_x,training_y
X1,X2 = np.meshgrid(np.arange(start = X_set[:,0].min()-1,stop = X_set[:,0].max()+1,step = 0.01),
                   np.arange(start = X_set[:,1].min()-1,stop = X_set[:,1].max()+1,step = 0.01)) 
plt.contourf(X1,X2 ,classifer_LR.predict(np.array([X1.ravel(),X2.ravel()]).T).reshape(X1.shape),alpha=0.75 ,cmap=ListedColormap(("red","green")))
plt.xlim(X1.min(),X1.max())
plt.ylim(X2.min(),X2.max())
for i ,j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j,0],X_set[y_set == j,1],c=ListedColormap(("red","green"))(i),label=j)

plt.title("Logistic Regression (Training set)")
plt.xlabel("Age")
plt.ylabel("Estimetes Salary")
plt.legend()
plt.show()
```

```python
# Testing dataset

from matplotlib.colors import ListedColormap
X_set , y_set = test_x,test_y
X1,X2 = np.meshgrid(np.arange(start = X_set[:,0].min()-1,stop = X_set[:,0].max()+1,step = 0.01),
                   np.arange(start = X_set[:,1].min()-1,stop = X_set[:,1].max()+1,step = 0.01)) 
plt.contourf(X1,X2 ,classifer_LR.predict(np.array([X1.ravel(),X2.ravel()]).T).reshape(X1.shape),alpha=0.75 ,cmap=ListedColormap(("red","green")))
plt.xlim(X1.min(),X1.max())
plt.ylim(X2.min(),X2.max())
for i ,j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j,0],X_set[y_set == j,1],c=ListedColormap(("red","green"))(i),label=j)

plt.title("Logistic Regression (Testing set)")
plt.xlabel("Age")
plt.ylabel("Estimetes Salary")
plt.legend()
plt.show()
```
