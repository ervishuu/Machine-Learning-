```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("Job_Exp.csv")
data.head(10)
```

```python
real_x  = data.iloc[: , [0]].values
real_y = data.iloc[: , 1].values
```

```python
from sklearn.model_selection import train_test_split
training_x ,test_x , training_y , test_y = train_test_split(real_x ,real_y , test_size = 0.33, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt_r = DecisionTreeRegressor()
dt_r.fit(training_x , training_y)
```

```python
dt_r.score(training_x,training_y)
```

```python
y_pred = dt_r.predict(test_x)
```

```python
print(y_pred)
```

```python
#graph
real_x_dt = np.arange(min(real_x),max(real_x),0.1)
real_x_dt = real_x_dt.reshape(len(real_x_dt),1)
# print(real_x_dt)
plt.scatter(real_x,real_y,color = "black")
plt.plot(real_x_dt,dt_r.predict(real_x_dt),color = "red")
plt.ylabel("Getting Job Chance %")
plt.xlabel("Year Of EXP")
plt.show()
```
